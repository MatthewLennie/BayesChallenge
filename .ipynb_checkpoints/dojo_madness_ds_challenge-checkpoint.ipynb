```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
# feel free to change the variable name to something you are comfortable with
dat = pd.read_csv("challenge_data.csv")
keys = ['game_time', 'winner', 'feature_1','feature_2', 'feature_3', 'feature_4']
```

# Something is wrong in the data?

```python
dat.dtypes
```

```python
#%% Investigate weirdness and remove the text value trap ;) 
how_many_weird = pd.to_numeric(dat['feature_4'], errors='coerce').isna().sum()

#ok only one weird. What is it?
weird_entry = dat['feature_4'][pd.to_numeric(dat['feature_4'], errors='coerce').isna()]
print(weird_entry)
#Just remove the single entry. 
dat.drop(weird_entry.index,inplace=True)

dat.loc[:,'feature_4'] = pd.to_numeric(dat.loc[:,'feature_4'], errors='coerce')
```

# Colinear Features

```python
#%% Look at the data in the simplest form. 

# Feature_4 is suspect
dat.corr()

# it has an outlier and seems to just be a copy of the game_time anyway. 
sns.pairplot(dat[keys])

# Just demonstrating the point a little further here to show the massive outlier
results = smf.ols('feature_4 ~ game_time',data=dat).fit()
sm.graphics.influence_plot(results)
```

```python
if 'feature_4' in dat.keys():
    dat.drop('feature_4',axis=1,inplace=True)
else:
    print("already removed")
```

dat


# Test Train Split and normalization

```python
X_train, X_test, y_train, y_test = train_test_split(dat[['game_time','feature_1','feature_2','feature_3']], dat['winner'], test_size=0.33, random_state=42)
#train scaler on the training data only, no data leakages. 
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Try fitting a few cheap classification models

In this quick analysis I will ignore the autocorrelation of the data

```python
#LDA is probably good considering that two of the features are highly correlated. 

clf1 = LinearDiscriminantAnalysis()
clf1.fit(X_train_scaled, y_train)
print("training {}".format(clf1.score(X_train_scaled,y_train)))
print("validation {}".format(clf1.score(X_test_scaled,y_test)))
```

```python
#%% Doesn't work really well. If you look at the scatter plots you can see why. the provided features don't really give good class separation

clf2 = LogisticRegression(random_state=0, solver='lbfgs').fit(X_train, y_train)
print(clf2.score(X_train_scaled,y_train))
print(clf2.score(X_test_scaled,y_test))
```

```python
#%% Random Forest Classifier
# nice way of combining the interaction of features. 
# max_depth = 4 optimizes bias/Variance Tradeoff 

clf3 = RandomForestClassifier(n_estimators=100, max_depth=4,random_state=0)
clf3.fit(X_train, y_train)

print(clf3.score(X_train, y_train))
print(clf3.score(X_test, y_test))
```

```python
# Quickly tried an ensemble hoping that there was a lack of correlation between the predictions 
# Wasn't really a success
eclf2 = VotingClassifier(estimators=[
        ('lda', clf1), ('lr', clf2), ('rf', clf3)],
        voting='soft')
eclf2 = eclf2.fit(X_train, y_train)
print(eclf2.score(X_train,y_train))
print(eclf2.score(X_test,y_test))
```

With my experiments, I got it up to 63% on the valdiation, I tend to think that it's not possible to squeeze much more out of this data. Mostly because the real feature that we have looks like this;

```python
dat.plot.scatter('feature_1','winner')
```

Feature 3 is next to useless as it is essentially sparse and feature 2 is more or less the same as feature 1 with different noise. 

The last thing to consider is that we have little data in the later stages of long games. i.e. it's a survival process meaning as time increases, only small amounts of game data will be availiable. I didn't bother adjusting anything for this effect with the simple models I implemented, but if you started doing time series models or something more complicated, then it should be taken into account.
